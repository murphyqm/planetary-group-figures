# Seaborn

Seaborn is a very aesthetically-pleasing plotting library, tailored for
statistical applications but also has lots of general functionality.

Look at the gallery [here](https://seaborn.pydata.org/examples/index.html).

Seaborn colour palettes: https://www.codecademy.com/articles/seaborn-design-ii



```python
import seaborn as sns
sns.set_theme() # reset to default

# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")

# Plot the responses for different events and regions
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

```python
sns.set_theme()
sns.set_theme(style="ticks")

df = sns.load_dataset("penguins")
sns.pairplot(df, hue="species")
```

Can change the colour scheme to one of the inbuilt palettes... deep, muted, pastel, bright, dark, and colorblind.

```python
sns.set_theme()
sns.set_theme(style="ticks")
sns.set_palette("pastel")

df = sns.load_dataset("penguins")
sns.pairplot(df, hue="species")
```

You can also pull in color brewer palettes... https://www.codecademy.com/articles/seaborn-design-ii

https://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3 - inspired by the research of the cartographer Cindy Brewer.


(I've used a sequential palette below just to show how to use it but obviously should pick a categorical
palette for this plot)

```python
sns.set_theme()
sns.set_theme(style="ticks")
custom_palette = sns.color_palette("RdPu", 3)
sns.set_palette(custom_palette)

df = sns.load_dataset("penguins")
sns.pairplot(df, hue="species")

```

Seaborn also comes with a series of inbuilt themes, not as many as matplotlib:
`darkgrid`, `whitegrid`, `dark`, `white`, and `ticks`.

```python
sns.set_theme() # reset to default
sns.set_palette("deep")
sns.set_theme(style="darkgrid")
sns.pairplot(df, hue="species")
```

Seaborn also has a useful functionality called context, with
four different options: in order of relative size,  paper,
notebook, talk, and poster.


```python
sns.set_theme()
sns.set_context("paper")

sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

```python
sns.set_theme()
sns.set_context("talk")

sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

```python
sns.set_theme()
sns.set_context("poster")

sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri, legend=False)
```

```python
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri, legend=False)
```

```python
sns.set_theme()
sns.set_context("paper")

sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```
