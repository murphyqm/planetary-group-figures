# More colormaps

Cmocean is a library of perceptually-uniform colormaps specifically designed for use
in oceanography, a bit more choice than the 4 perceptually uniform matplotlib ones.




```python
import cmocean
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Always useful to remember to reset to default:
plt.style.use('default')

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.8, cmap=cmocean.cm.thermal)
plt.colorbar()  # show color scale
```
