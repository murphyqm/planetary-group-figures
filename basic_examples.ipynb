## matplotlib

### The basics

Scatterplot demo from [matplotlib](https://matplotlib.org/stable/gallery/lines_bars_and_markers/scatter_masked.html#sphx-glr-gallery-lines-bars-and-markers-scatter-masked-py)

Scatterplot where certain values are masked and given different shape symbols,
with size defined by variable value (in this case, random).

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)


N = 100
r0 = 0.6
x = 0.9 * np.random.rand(N)
y = 0.9 * np.random.rand(N)
area = (20 * np.random.rand(N))**2  # 0 to 10 point radii
c = np.sqrt(area)
r = np.sqrt(x ** 2 + y ** 2)
area1 = np.ma.masked_where(r < r0, area)
area2 = np.ma.masked_where(r >= r0, area)
plt.scatter(x, y, s=area1, marker='^', c=c)
plt.scatter(x, y, s=area2, marker='o', c=c)
# Show the boundary between the regions:
theta = np.arange(0, np.pi / 2, 0.01)
plt.plot(r0 * np.cos(theta), r0 * np.sin(theta))

plt.show()
```

# Scatter plot with histograms

Based on the example [here](https://matplotlib.org/stable/gallery/lines_bars_and_markers/scatter_hist.html#sphx-glr-gallery-lines-bars-and-markers-scatter-hist-py)




```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# some random data
x = np.random.randn(1000)
y = np.random.randn(1000)

def scatter_hist(x, y, ax, ax_histx, ax_histy):
    # no labels
    ax_histx.tick_params(axis="x", labelbottom=False)
    ax_histy.tick_params(axis="y", labelleft=False)

    # the scatter plot:
    ax.scatter(x, y)

    # now determine nice limits by hand:
    binwidth = 0.25
    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    lim = (int(xymax/binwidth) + 1) * binwidth

    bins = np.arange(-lim, lim + binwidth, binwidth)
    ax_histx.hist(x, bins=bins)
    ax_histy.hist(y, bins=bins, orientation='horizontal')
```

```python
# definitions for the axes
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
spacing = 0.005


rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom + height + spacing, width, 0.2]
rect_histy = [left + width + spacing, bottom, 0.2, height]

# start with a square Figure
fig = plt.figure(figsize=(8, 8))

ax = fig.add_axes(rect_scatter)
ax_histx = fig.add_axes(rect_histx, sharex=ax)
ax_histy = fig.add_axes(rect_histy, sharey=ax)

# use the previously defined function
scatter_hist(x, y, ax, ax_histx, ax_histy)

plt.show()
```

## Making stuff look a bit nicer

Matplotlib has a bunch of predefined available styles or themes that can
be edited by hand to suit, see more [here](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html).

Also can assign different [colourmaps](https://matplotlib.org/stable/tutorials/colors/colormaps.html).
These are sorted into groups, including a selection of "perceptually uniform sequential
colourmaps".

Viridis is the default colourmap...



```python

# Always useful to remember to reset to default:
plt.style.use('default')

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()  # show color scale
```

As well as changing the colurmap, axes, etc by hand, can also use matplotlib styles.



```python
# What styles are available?
print(plt.style.available)
```

```python
def basic_plot():
    rng = np.random.RandomState(0)
    sizes = 500 * rng.rand(10)
    x1 = rng.randn(10)
    y1 = rng.randn(10)

    rng2 = np.random.RandomState(2)
    sizes2 = 500 * rng2.rand(10)
    x2 = rng2.randn(10)
    y2 = rng2.randn(10)

    rng3 = np.random.RandomState(3)
    sizes3 = 500 * rng3.rand(10)
    x3 = rng3.randn(10)
    y3 = rng3.randn(10)

    plt.scatter(x1, y1, s=sizes, alpha=0.8)
    plt.scatter(x2, y2, s=sizes2, alpha=0.8)
    plt.scatter(x3, y3, s=sizes3, alpha=0.8)

    plt.xlabel("This is the x-axis")
    plt.ylabel("The y-axis")
    plt.title("Wow! What a plot.")

    plt.show()
```

```python
#Always useful to remember to reset to default:
plt.style.use('default')

basic_plot()
```

```python
#Always useful to remember to reset to default:
plt.style.use('default')
plt.style.use('seaborn-pastel')
basic_plot()
```

```python
#Always useful to remember to reset to default:
plt.style.use('default')
plt.style.use('ggplot')
basic_plot()

```

```python
#Always useful to remember to reset to default:
plt.style.use('default')
plt.style.use('bmh')
basic_plot()
```
